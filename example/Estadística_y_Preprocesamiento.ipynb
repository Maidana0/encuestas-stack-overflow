#**Fundamentos de Estadística**
**Estadística Descriptiva:** Concepto y Relevancia

La estadística descriptiva es una rama de la estadística que se encarga de recopilar, organizar, describir y analizar un conjunto de datos con el fin de extraer información relevante.

Su objetivo principal es ofrecer una visión general y comprensible de los datos, lo que permite identificar patrones, tendencias y relaciones dentro del conjunto de datos.

La importancia de la estadística descriptiva radica en su capacidad para proporcionar un resumen claro y conciso de los datos, lo cual es esencial para la toma de decisiones informadas. Al utilizar medidas de resumen como la media, mediana, moda, varianza y desviación estándar, podemos obtener una comprensión más profunda de la distribución de los datos y de las características principales del conjunto de datos analizado.

#**Introducción al Análisis Exploratorio de Datos (EDA)**

El Análisis Exploratorio de Datos (EDA, por sus siglas en inglés) es una fase crítica en el análisis de datos que se centra en descubrir patrones, detectar anomalías, probar hipótesis y verificar supuestos mediante el uso de estadísticas descriptivas y herramientas de visualización. La filosofía detrás del EDA es acercarse a los datos sin suposiciones previas, permitiendo así una exploración abierta que puede revelar aspectos inesperados o interesantes de los datos.

El EDA no solo ayuda a entender la estructura y características de los datos, sino que también es fundamental para preparar los datos para análisis más complejos, como la modelización predictiva. Mediante el uso de gráficos, tablas y medidas de resumen, el EDA permite a los analistas de datos tomar decisiones informadas sobre la adecuación de los datos para los análisis posteriores.

Resumen

La estadística descriptiva es crucial en el análisis de datos, ya que permite resumir y comprender los datos de manera efectiva. Es una herramienta clave para el Análisis Exploratorio de Datos (EDA), que se utiliza para explorar y analizar los datos en profundidad, ayudando a revelar patrones, anomalías y otras características importantes antes de realizar análisis más complejos.

Tipos de Variables
En este apartado sobre los diferentes tipos de variables y su importancia en el análisis de datos.
*   **Cualitativas (Categóricas):** Representan cualidades o categorías.
    *   **Nominal:** Categorías sin orden (ej: género, color).
    *   **Ordinal:** Categorías con un orden o jerarquía (ej: nivel educativo, satisfacción).
*   **Cuantitativas (Numéricas):** Representan cantidades medibles.
    *   **Discretas:** Valores contables, a menudo enteros (ej: número de hijos).
    *   **Continuas:** Pueden tomar cualquier valor dentro de un rango (ej: altura, temperatura).

#**Medidas de Tendencia Central**
**Medidas Cuantitativas**

#**Media**

La media es una medida de tendencia central que se calcula sumando todos los valores de un conjunto de datos y dividiendo esta suma por la cantidad de valores. Es uno de los indicadores más utilizados para resumir un conjunto de datos, ya que ofrece una idea general del valor promedio.

**Uso de la Media:** La media es útil cuando se necesita un valor representativo del conjunto de datos. Sin embargo, puede ser influenciada por valores atípicos (outliers), lo que puede distorsionar su interpretación.

#**Mediana**
La mediana es el valor que ocupa la posición central en un conjunto de datos ordenados de menor a mayor. Si el número de observaciones es impar, la mediana es el valor en la posición central. Si es par, la mediana es el promedio de los dos valores centrales.

**Uso de la Mediana:** La mediana es especialmente útil en distribuciones asimétricas o cuando existen valores atípicos, ya que no se ve afectada por estos como la media.

#**Moda**
La moda es el valor que aparece con mayor frecuencia en un conjunto de datos. Un conjunto de datos puede tener una única moda (unimodal), más de una moda (multimodal) o ninguna moda si todos los valores son únicos.

**Uso de la Moda:** La moda se utiliza principalmente en variables cualitativas o en situaciones donde se requiere conocer el valor más común dentro del conjunto de datos.

## Medidas de Tendencia Central para las Estadísticas de Pokémon

Vamos a calcular la media, mediana y moda para una estadística numérica, como el 'Attack', en el dataset de Pokémon.

```python
# MEDIA para 'Attack'
print("Media de Attack:", df_pokemon['Attack'].mean())

# MEDIANA para 'Attack'
print("Mediana de Attack:", df_pokemon['Attack'].median())

# MODA para 'Attack'
print("Moda de Attack:", df_pokemon['Attack'].mode())
```

## Gráfico de Distribución de Attack en Pokémon con Media y Mediana

Vamos a visualizar la distribución de la estadística 'Attack' en el dataset de Pokémon y marcar la media y la mediana en el gráfico.

```python
# Calcular media y mediana para 'Attack'
media_attack = df_pokemon['Attack'].mean()
mediana_attack = df_pokemon['Attack'].median()

# Crear el gráfico de distribución
plt.figure(figsize=(10, 6))
sns.histplot(df_pokemon['Attack'], bins=20, kde=True, color='salmon', edgecolor='black')

# Añadir líneas para la media y la mediana
plt.axvline(media_attack, color='red', linestyle='--', linewidth=2, label=f'Media: {media_attack:.2f}')
plt.axvline(mediana_attack, color='green', linestyle='-', linewidth=2, label=f'Mediana: {mediana_attack:.2f}')

# Configurar títulos y etiquetas
plt.title('Distribución de Attack en Pokémon')
plt.xlabel('Attack')
plt.ylabel('Frecuencia')
plt.legend()

# Mostrar el gráfico
plt.show()
```

#**Varianza**
La varianza mide la dispersión de los datos respecto a la media. Se calcula promediando el cuadrado de las diferencias entre cada valor y la media. La varianza ofrece una idea de cuán dispersos están los datos; una varianza alta indica que los datos están muy dispersos, mientras que una varianza baja indica que los datos están más concentrados alrededor de la media.


**Uso de la Varianza:** La varianza es fundamental para entender la variabilidad dentro de un conjunto de datos, y es una base para el cálculo de otras medidas como la desviación estándar.

#**Desviación Estándar**
La desviación estándar es la raíz cuadrada de la varianza y proporciona una medida de dispersión en las mismas unidades que los datos originales. Es una medida más intuitiva que la varianza porque refleja directamente la magnitud de las desviaciones respecto a la media.

Uso de la Desviación Estándar: La desviación estándar es ampliamente utilizada para entender la dispersión de los datos y es fundamental en muchas aplicaciones estadísticas, incluyendo la evaluación de la volatilidad en mercados financieros o la variabilidad en procesos de producción.

#**Cuartiles y Percentiles**
Los cuartiles y percentiles son medidas que dividen un conjunto de datos ordenados en partes iguales, proporcionando información detallada sobre la distribución de los datos. Los cuartiles dividen los datos en cuatro partes iguales: el primer cuartil (Q1) marca el 25% de los datos más bajos, el segundo cuartil (Q2), que corresponde a la mediana, divide el conjunto en la mitad, y el tercer cuartil (Q3) marca el 75% de los datos más bajos. Los percentiles, por su parte, dividen los datos en 100 partes iguales, y son especialmente útiles para interpretar la posición relativa de un valor dentro de un conjunto de datos. Por ejemplo, si un valor se encuentra en el percentil 90, significa que el 90% de los valores en el conjunto son menores o iguales a él.

**Uso de Cuartiles y Percentiles:** Estas medidas son fundamentales para análisis en los que se requiere comprender la dispersión y distribución de los datos más allá de la media y la mediana, permitiendo identificar rangos y establecer comparaciones entre diferentes subconjuntos de datos.



## Medidas de Variabilidad o Dispersión para las Estadísticas de Pokémon

Vamos a calcular algunas medidas de variabilidad para la estadística 'Attack' en el dataset de Pokémon para entender su dispersión.

```python
# VARIANZA para 'Attack'
print("Varianza de Attack:", df_pokemon['Attack'].var())

# DESVIACIÓN ESTÁNDAR para 'Attack'
print("Desviación estándar de Attack:", df_pokemon['Attack'].std())

# RANGO para 'Attack'
print("Rango de Attack:", df_pokemon['Attack'].max() - df_pokemon['Attack'].min())

# COEFICIENTE DE VARIABILIDAD para 'Attack'
cv_attack = df_pokemon['Attack'].std() / df_pokemon['Attack'].mean()
print("Coeficiente de variabilidad de Attack:", cv_attack)

# Interpretación sugerida para el Coeficiente de Variabilidad
if cv_attack < 0.05:
    print("CV < 5%: La variable 'Attack' tiene baja variabilidad relativa.")
else:
    print("CV > 5%: La variable 'Attack' tiene alta variabilidad relativa.")
```

#**Medidas Cualitativas**
Medidas de Resumen para Variables Cualitativas

Las variables cualitativas, también conocidas como variables categóricas, son aquellas que describen cualidades o características y no pueden ser medidas numéricamente de manera directa. Ejemplos comunes de variables cualitativas incluyen el género, el estado civil, la categoría de un producto, entre otros. Para resumir este tipo de datos, se utilizan principalmente dos medidas: el conteo de observaciones y la moda.

**Conteo de Observaciones**

El conteo de observaciones es una medida básica pero poderosa para variables cualitativas. Consiste en contar la cantidad de veces que cada categoría aparece en el conjunto de datos. Este conteo permite obtener una visión clara de la distribución de las categorías dentro del conjunto de datos.

Uso del Conteo de Observaciones: Esta medida es útil para entender la frecuencia de cada categoría y para comparar cómo se distribuyen las diferentes categorías en el conjunto de datos. Por ejemplo, en un análisis de mercado, el conteo de observaciones puede mostrar cuántos clientes pertenecen a cada segmento de mercado.

**Cálculo de la Moda**

La moda es la categoría que aparece con mayor frecuencia en un conjunto de datos cualitativos. A diferencia de la media y la mediana, que son más comunes en variables cuantitativas, la moda es la medida de tendencia central más adecuada para variables cualitativas. Un conjunto de datos puede ser unimodal (una sola moda), bimodal (dos modas) o multimodal (más de dos modas).

Uso de la Moda: La moda es particularmente útil cuando se quiere identificar la categoría más representativa o común en el conjunto de datos. Por ejemplo, en un estudio de preferencias de productos, la moda indicaría cuál es el producto preferido por la mayoría de los encuestados.

#**Distribuciones de Variables**
**Distribuciones Uniformes**  

Una distribución uniforme es aquella en la que todos los valores posibles dentro de un rango tienen la misma probabilidad de ocurrir. En otras palabras, cada resultado es igualmente probable. Esta distribución es a menudo representada por una forma rectangular en un histograma, donde la altura de cada barra es igual, reflejando que todas las categorías tienen la misma frecuencia.

Ejemplo de Distribución Uniforme: Un ejemplo clásico de distribución uniforme es el lanzamiento de un dado justo, donde cada número del 1 al 6 tiene la misma probabilidad de salir (1/6).

**Distribuciones Normales**

La distribución normal, también conocida como la campana de Gauss, es una de las distribuciones más importantes en estadística. Se caracteriza por ser simétrica respecto a su media y tener una forma de campana. La mayor parte de los datos se agrupa alrededor de la media, y la probabilidad disminuye a medida que nos alejamos de la media en ambos sentidos. La media, mediana y moda coinciden en el centro de la distribución.

Ejemplo de Distribución Normal: Un ejemplo típico de una distribución normal es la estatura de los adultos de una población, donde la mayoría de las personas tienen estaturas cercanas al promedio, y hay pocas personas muy bajas o muy altas.

Diferencias entre Distribuciones Uniformes y Normales

La principal diferencia entre una distribución uniforme y una normal es la forma de la distribución de los datos. Mientras que la distribución uniforme muestra que todos los resultados son igualmente probables, la distribución normal indica que los resultados cercanos a la media son más probables, y los extremos son menos probables.

Forma: La distribución uniforme es rectangular, mientras que la distribución normal tiene forma de campana.

Simetría: Aunque ambas distribuciones pueden ser simétricas, la distribución normal tiene una concentración de datos alrededor de la media, mientras que la uniforme no tiene tal concentración.

#**Interpretación de Histogramas**

Un histograma es una representación gráfica de la distribución de un conjunto de datos. En un histograma, el eje horizontal (X) muestra las categorías o intervalos de los datos, y el eje vertical (Y) muestra la frecuencia o número de observaciones en cada categoría.

Interpretación de Histogramas en Distribuciones Uniformes: En un histograma de una distribución uniforme, se espera ver barras de igual altura, indicando que cada intervalo tiene aproximadamente el mismo número de observaciones.

Interpretación de Histogramas en Distribuciones Normales: En un histograma de una distribución normal, se espera ver una forma de campana, con las barras centrales más altas, que representan la mayor concentración de datos alrededor de la media, y barras más bajas a medida que nos alejamos de la media.

#**Interpretación de la Correlación Lineal y el Coeficiente de Correlación**
La correlación lineal se refiere a la relación de linealidad entre dos variables. Esta relación se mide a través del coeficiente de correlación, que toma valores entre -1 y 1:

Un coeficiente de correlación cercano a 1 indica una fuerte correlación positiva, lo que significa que, en general, a mayores valores de una variable le corresponden mayores valores de la otra.

Un coeficiente cercano a -1 indica una fuerte correlación negativa, lo que significa que a mayores valores de una variable le corresponden menores valores de la otra.

Un coeficiente cercano a 0 sugiere que no hay una correlación lineal fuerte entre las dos variables, es decir, los cambios en una variable no están asociados de manera consistente con cambios en la otra.

Es crucial recordar que la correlación no implica causalidad. Aunque dos variables puedan mostrar una relación aparente en un diagrama de dispersión, esto no significa necesariamente que una cause la variación en la otra. Para hacer afirmaciones causales, se requiere un análisis más profundo, como la regresión y otros métodos estadísticos más avanzados.

Los diagramas de dispersión son una herramienta poderosa para descubrir relaciones y patrones ocultos en los datos, pero siempre deben interpretarse con cuidado y en el contexto adecuado.

## Medidas de Forma para la Estadística Attack en Pokémon

Analicemos la forma de la distribución de la estadística 'Attack' en el dataset de Pokémon.

```python
# Calcular la asimetría (skewness) para 'Attack'
asimetria_attack = df_pokemon['Attack'].skew()
print("Asimetría (skewness) de Attack:", asimetria_attack)

# Calcular media y moda para 'Attack'
media_attack = df_pokemon['Attack'].mean()
# El método .mode() puede devolver múltiples valores si hay un empate, tomamos el primero.
moda_attack = df_pokemon['Attack'].mode()[0]

# Interpretación de la asimetría
print("\nInterpretación de la asimetría:")
if moda_attack > media_attack:
    print("La distribución de Attack es asimétrica negativa (cola a la izquierda).")
elif moda_attack == media_attack:
    print("La distribución de Attack es simétrica.")
else:
    print("La distribución de Attack es asimétrica positiva (cola a la derecha).")

# Gráfico: Histograma con líneas de media y moda para 'Attack'
plt.figure(figsize=(10, 6))
sns.histplot(df_pokemon['Attack'], bins=20, kde=True, color='skyblue', edgecolor='black')

# Añadir líneas para la media y la moda
plt.axvline(media_attack, color='red', linestyle='--', linewidth=2, label=f'Media: {media_attack:.2f}')
plt.axvline(moda_attack, color='green', linestyle='-', linewidth=2, label=f'Moda: {moda_attack:.2f}')

# Configurar títulos y etiquetas
plt.title('Distribución de Attack en Pokémon con Media y Moda')
plt.xlabel('Attack')
plt.ylabel('Frecuencia')
plt.legend()

# Mostrar el gráfico
plt.show()
```

La asimetría (skewness) para la estadística 'Attack' indica si su distribución está sesgada hacia valores altos (asimetría positiva) o bajos (asimetría negativa). El gráfico de histograma con las líneas de la media y la moda nos ayuda a visualizar esta forma de la distribución.

En este caso, la asimetría calculada y la posición relativa de la media y la moda nos permiten determinar la dirección del sesgo en la distribución de los valores de 'Attack' en el dataset de Pokémon.

#**Detección y Manejo de Outliers**
**Detección de Outliers**

Los outliers, o valores atípicos, son observaciones que se encuentran a una distancia anormal del resto de los datos en un conjunto. Identificar y tratar estos outliers es crucial para evitar que distorsionen los resultados del análisis estadístico o de los modelos predictivos. Una de las técnicas más efectivas y visualmente intuitivas para detectar outliers es el uso de diagramas de caja y bigotes (boxplots).

1. Identificación de Outliers con Diagramas de Caja y Bigotes

Un diagrama de caja y bigotes es una herramienta gráfica que permite visualizar la distribución de los datos y resaltar posibles outliers. Este diagrama se construye a partir de los cuartiles de los datos:

Caja (Box): Representa el rango intercuartil (IQR), que es la diferencia entre el tercer cuartil (Q3) y el primer cuartil (Q1). La caja se extiende desde Q1 hasta Q3 y contiene el 50% central de los datos.

Bigotes (Whiskers): Se extienden desde los extremos de la caja hasta el valor mínimo y máximo dentro de 1.5 veces el IQR desde Q1 y Q3, respectivamente.

Outliers: Cualquier punto de datos que caiga fuera de los límites de los bigotes se considera un outlier.

2. Tratamiento de Outliers

Una vez que se han identificado los outliers, es esencial decidir cómo tratarlos, lo que dependerá del contexto y del impacto que estos valores puedan tener en el análisis:

Eliminación de Outliers: En casos donde los outliers se deben a errores de medición o entrada de datos, es común eliminarlos para evitar que distorsionen los resultados.

Transformación de Datos: Los outliers pueden ser suavizados o transformados utilizando técnicas como la transformación logarítmica, que reduce su impacto en el análisis.

Imputación de Valores: Si los outliers son representativos de datos válidos pero extremos, en algunos casos se pueden imputar utilizando el valor del cuartil más cercano o alguna otra técnica de sustitución.

Análisis de Sensibilidad: En ciertos contextos, puede ser útil realizar un análisis de sensibilidad con y sin outliers para entender su impacto en el modelo o en las conclusiones del análisis.

#**Conclusión**

El uso de diagramas de caja y bigotes es una técnica eficaz para la identificación de outliers en un conjunto de datos. Una vez identificados, el tratamiento de estos outliers debe ser considerado cuidadosamente, ya que pueden tener un impacto significativo en los resultados del análisis y en la precisión de los modelos predictivos. La elección del método de tratamiento dependerá del origen de los outliers y del contexto del análisis

## Calcular el límite superior de outliers

### Subtask:
Determinar el límite superior para identificar outliers en la estadística 'Attack' usando el método del rango intercuartílico (IQR).


**Reasoning**:
Calcular Q1, Q3, IQR y el límite superior para identificar outliers en la columna 'Attack' según el método IQR, y luego imprimir el límite superior.



```python
# 1. Calcula el primer cuartil (Q1) de la columna 'Attack'.
q1_attack = df_pokemon['Attack'].quantile(0.25)

# 2. Calcula el tercer cuartil (Q3) de la columna 'Attack'.
q3_attack = df_pokemon['Attack'].quantile(0.75)

# 3. Calcula el Rango Intercuartílico (IQR).
iqr_attack = q3_attack - q1_attack

# 4. Calcula el límite superior para la detección de outliers.
upper_limit_attack = q3_attack + 1.5 * iqr_attack

# 5. Imprime el valor calculado del límite superior.
print(f"Límite superior para outliers en Attack (método IQR): {upper_limit_attack:.2f}")
```

## Filtrar el dataset sin outliers

### Subtask:
Crear un nuevo DataFrame excluyendo los Pokémon cuyos valores de 'Attack' superan el límite superior calculado.


**Reasoning**:
Crear un nuevo DataFrame excluyendo las filas donde el valor de 'Attack' sea mayor que el límite superior calculado, luego mostrar las primeras filas y la forma del nuevo DataFrame para verificar el resultado.

```python
# 1. Crea un nuevo DataFrame excluyendo los Pokémon con 'Attack' mayor que el límite superior calculado.
df_pokemon_no_outliers = df_pokemon[df_pokemon['Attack'] <= upper_limit_attack].copy()

# 2. Muestra las primeras filas del nuevo DataFrame para verificar el filtrado.
print("DataFrame de Pokémon sin outliers en Attack:")
display(df_pokemon_no_outliers.head())

# 3. Imprime la forma del nuevo DataFrame para ver cuántas filas se eliminaron.
print("\nForma del DataFrame sin outliers en Attack:")
print(df_pokemon_no_outliers.shape)
```

## Calcular la media de attack por type 1 (con outliers)

### Subtask:
Calcular la media de la estadística 'Attack' para cada 'Type 1' utilizando el DataFrame original.


**Reasoning**:
Calcular la media de la columna 'Attack' agrupada por la columna 'Type 1' del DataFrame original y mostrar los resultados.



```python
# Calcula la media de la columna 'Attack' agrupada por la columna 'Type 1' del DataFrame original.
mean_attack_by_type_with_outliers = df_pokemon.groupby('Type 1')['Attack'].mean()

# Muestra las medias calculadas.
print("Media de Attack por 'Type 1' (con outliers):")
display(mean_attack_by_type_with_outliers)
```

Combinar las series de media de ataque con y sin outliers, renombrar columnas, derretir el dataframe para graficar, y luego crear un gráfico de barras para comparar las medias por Type 1, rotando las etiquetas del eje x para legibilidad y estableciendo el tamaño de la figura, título y etiquetas.
